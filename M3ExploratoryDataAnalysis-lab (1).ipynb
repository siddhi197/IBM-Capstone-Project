## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
plt.figure(figsize=(10, 6))

# Create a box plot for the 'ConvertedComp' column
sns.boxplot(x=df['ConvertedComp'], color='blue')

# Add titles and labels
plt.title('Box Plot of Converted Annual Salary (in USD)', fontsize=16)
plt.xlabel('Converted Annual Salary (USD)', fontsize=14)

# Show the plot
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)

IQR = Q3 - Q1

print(f"Interquartile Range (IQR) for 'ConvertedComp': {IQR}")
```

Find out the upper and lower bounds.

```python
# your code goes here
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print(f"Lower bound for outliers: {lower_bound}")
print(f"Upper bound for outliers: {upper_bound}")
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound) ]

print(f"Number of outliers: {len(outliers)}")
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_clean = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
print(f"Original DataFrame shape: {df.shape}")
print(f"New DataFrame shape after removing outliers: {df_clean.shape}")
```
